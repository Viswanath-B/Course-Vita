```python
import pandas as pd
import numpy as np
df = pd.read_csv("/content/diabetes_dataset.csv")
print("Preview of data:")
print(df.head())
print("\nData Info:")
print(df.info())
```

```python
dummy = df.drop_duplicates()
print("\nAfter removing duplicates:", dummy.shape)
```

```python
rows, cols = df.shape
print("Rows:", rows)
print("Columns:", cols)
```

```python
activity_count = df.groupby(['Sex', 'Physical_Activity_Level']).size()
print(activity_count.to_string())
```

```python
male_count = np.sum(df['Sex'] == 'Male')
female_count = np.sum(df['Sex'] == 'Female')
print("Males:", male_count)
print("Females:", female_count)
```

```python
remove_empty_cells = df.dropna()
print(remove_empty_cells.to_string())
```
