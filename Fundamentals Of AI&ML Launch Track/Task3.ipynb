```python
ages = df['age'].dropna()
average_age = ages.mean()
middle_age = ages.median()
most_common_age = ages.mode()[0]
print("Mean (Average Age):", round(average_age, 2))
print("Median (Middle Age):", round(middle_age, 2))
print("Mode (Most Common Age):", round(most_common_age, 2))
```

```python
def age_group(age):
    if age < 18:
        return "Children"
    elif age < 60:
        return "Adult"
    else:
        return "Old"
df_clean = df[['age', 'bp']].dropna().copy()
df_clean['age_group'] = df_clean['age'].apply(age_group)
plt.figure()
for group in ['Children', 'Adult', 'Old']:
    subset = df_clean[df_clean['age_group'] == group]
    plt.scatter(subset['age'], subset['bp'], label=group)
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.title("Age vs BP by Age Group")
plt.legend()
plt.show()
```
